```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 3
x = np.arange(-n, n, 0.01)
tanh_x = np.tanh(x)
sinh_x = np.sinh(x)
cosh_x = np.cosh(x)
plt.plot(x, tanh_x, label='y = tanh(x)')
plt.plot(x, sinh_x, label='y = sinh(x)')
plt.plot(x, cosh_x, label='y = cosh(x)')
plt.legend()
plt.grid(True)
plt.ylim(-2*n, 2*n)
%matplotlib inline
```
